```python
class RelatorioDeAcoes():
    def __init__(self):
        self.nome_acao = input('Digite o nome da ação: ')
        self.data_inicial = input('Digite a data inicial no seguinte formato("AAAA-MM-DD"): ')
        self.data_final = input('Digite a data final de busca no seguinte formato("AAAA-MM-DD"): ')
        self.destinatario = 'ariewl.marcelino9904@gmail.com'
        self.assunto = 'Analise de ações de 2024'
        self.nome_usuario = 'Paulo Silva'
        self.nome_gestor = input('Digite o nome do Gestor para quem vai enviar: ')

    # Métodos
    # Listar informações da ações:
    def lista_de_acoes(self):
        acoes = yfinance.Ticker(self.nome_acao).history(start=self.data_inicial,end=self.data_final)
        return acoes
    
    #Grafico de fechamento da ação
    def fechamento_da_acao(self):
        fechamento = self.lista_de_acoes().Close
        return fechamento.plot()
    
    # Cotação maxima da ação
    def contacao_maxima(self):
        contacao_max = round(fechamento_da_acao.max(),2)
        print(f'{contacao_max}')

    # Valor minimo da ação
    def cotacao_minima(self):
        cotacao_min = round(fechamento_da_acao.min(),2)
        print(f'{cotacao_min}')

    # Valor Total da cota
    def valor_medio(self):
        valor_med = round(fechamento_da_acao.mean(),2)
        print(f'{valor_med}')

relatorio = RelatorioDeAcoes()

```

```python
class EnviarEmailAoUsuario():
    def __init__(self):
        super().__init__()
        self.email = 'ariewl.marcelino9904@gmail.com'
        self.assunto = 'Analises do projeto 2020'
        self.colar_mensagem = f"""
                        Prezado Gestor,
                        Segue a analise das ações solicitadas:
                        Cotação minima: R$XXX Cotação maxima: R$XXX
                        Valor médio: R$XXX
                        Atenciosamente,
                        Paulo Silva
                        """
        self.abrir_web = webbrowser.open('www.gmail.com')
        self.clicar_em_escrever = pyautogui.click(x=2069, y=157)
        self.clicar_no_campo_destinatario = pyautogui.click(x=2495, y=356)
        self.copiar_email_destinatario = pyperclip.copy(self.email)
        self.colar = pyautogui.hotkey('ctrl','v')
        self.tab = pyautogui.hotkey('TAB')
        self.copiar_assunto = pyperclip.copy(self.assunto)
        self.copiar_mensagem = pyperclip.copy(self.colar_mensagem)
        self.clique_no_botão_enviar = pyautogui.click(x=2484, y=738)

    def abrir_navegador(self):
        #Abrir navegador:
        self.abrir_web
        time.sleep(3)
        print('abriu navegador')
    
        # #Espere 3s e clique no botão escrever:
        pyautogui.PAUSE = 3
        self.clicar_em_escrever
        print('clicou em escrever mensagem !')

    def colar_o_email_do_destinatario(self):
        #Colar o e-mail do destinatario
        self.clicar_no_campo_destinatario
        time.sleep(3)
        #para_destinatario
        self.copiar_email_destinatario
        self.colar
        print('adicionou e-mail do destinatario !')
        time.sleep(2)
        self.tab
    
    def assunto(self):
        self.copiar_assunto
        self.colar
        print('adicionou assunto !')
        time.sleep(2)
        self.tab
    
    def mensagem(self):
        # #Mensagem
        self.copiar_mensagem
        self.colar
        print('adicionou mensagem !')
        time.sleep(3)
    
    def enviar(self):
        #Clicar no enviar
        self.clique_no_botão_enviar
        print('Mensagem enviada!')

```
